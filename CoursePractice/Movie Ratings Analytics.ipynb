```python
#Stacked Histograms
```

```python
#Create Filter
# movies.Genre == "Drama"  
# movies[movies.Genre == "Drama"].BudgetMillions #Square Bracket!!!!

plt.hist(movies[movies.Genre == "Action"].BudgetMillions, bins = 15)
plt.hist(movies[movies.Genre == "Drama"].BudgetMillions, bins = 15)
plt.hist(movies[movies.Genre == "Thriller"].BudgetMillions, bins = 15)
plt.show()
```

```python
plt.hist([movies[movies.Genre == "Action"].BudgetMillions,\
          movies[movies.Genre == "Drama"].BudgetMillions,\
          movies[movies.Genre == "Thriller"].BudgetMillions,\
          movies[movies.Genre == "Comedy"].BudgetMillions
          ],\
            bins = 15, stacked =  True
        )
plt.show()
```

```python
for gen in movies.Genre.cat.categories:
    print(gen)
```

```python
list1 = list()
mylabels = list()

for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)
h = plt.hist(list1, bins = 30, stacked = True, rwidth=1, label=mylabels)
plt.legend()
plt.show()

#     plt.hist(movies[movies.Genre == gen].BudgetMillions,
#             bins = 15, stacked =  True
#         )
# plt.show() 
```

```python
#KDE Plot
```

```python
vis1 = sns.lmplot(data = movies, x = "CriticRating",\
                 y = "AudienceRating",\
                 fit_reg=False, hue = "Genre",\
                 height= 6, aspect=1)
```

```python
#Define what's x & y stand for 
k1 = sns.kdeplot(x = movies.CriticRating, \
                 y = movies.AudienceRating,\
                shade = True, shade_lowest = False, \
                cmap = "Reds") 

#TIP: make the plot smoother
k1b = sns.kdeplot(x = movies.CriticRating, \
                 y = movies.AudienceRating,\
                cmap = "Reds") 
```
