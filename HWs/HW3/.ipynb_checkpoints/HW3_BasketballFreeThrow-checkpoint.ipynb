```python
#PLOT1: Free Throw Attempts Per Game
def FTAperGame(data, playerlist = Players):
    Col = {"KobeBryant":"Black","JoeJohnson":"Red","LeBronJames":"Green","CarmeloAnthony":"Blue","DwightHoward":"Magenta","ChrisBosh":"Black","ChrisPaul":"Red","KevinDurant":"Green","DerrickRose":"Blue","DwayneWade":"Magenta"}
    Mrk = {"KobeBryant":"s","JoeJohnson":"o","LeBronJames":"^","CarmeloAnthony":"D","DwightHoward":"s","ChrisBosh":"o","ChrisPaul":"^","KevinDurant":"D","DerrickRose":"s","DwayneWade":"^"}
    for name in playerlist:
        plt.plot(data[Players_dict[name]]/Games[Players_dict[name]], c=Col[name], ls='--', marker=Mrk[name], ms=7, label=Players[Players_dict[name]])
   
    plt.title("Free Throw Attempts Per Game")
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)),Seasons) #, rotation = 'vertical'
    
    plt.xlabel("Seasons")
    plt.ylabel("Free Attempts")
    plt.show()
        
```
